```python
import json
import pandas as pd
```

```python
with open('../data/top18_teams_dec_2021.json', 'r', encoding='utf-8') as team_file:
    data_teams = json.load(team_file)
```

```python
teams_data =  pd.read_csv('../data/team_stats.csv')
teams_data
```

```python
dados = ['country','total_maps','kd_diff','kd','rating']

times = []

for team in data_teams:
    times.append(team['name'])
    if team['name'] == 'MOUZ':
        data = teams_data[teams_data['name']=='mousesports']
    else:
        data = teams_data[teams_data['name']==team['name']]

    for dado in dados:
        team[dado] = str(data[dado].iloc[0])

times
```

```python
teams_data[teams_data['name']=='mousesports']['country'].iloc[0]
```

```python
with open('../data/top18_teams.json', 'w', encoding='utf-8') as new_teams_file:
    json.dump(data_teams, new_teams_file)
```

```python
with open('../data/top20_players_2020.json', 'r', encoding='utf-8') as player_file:
    data_players = json.load(player_file)
```

```python
players_dataset =  pd.read_csv('../data/player_stats.csv')
players_dataset
```

```python
dados = ['country','total_maps','total_rounds','kd_diff','kd','rating']

jogadores = []

for player in data_players:
    jogadores.append(player['nick'])
    data = players_dataset[players_dataset['name']==player['nick']]

    for dado in dados:
        player[dado] = str(data[dado].iloc[0])
jogadores
```

```python
with open('../data/top20_players.json', 'w', encoding='utf-8') as new_players_file:
    json.dump(data_players, new_players_file)
```
